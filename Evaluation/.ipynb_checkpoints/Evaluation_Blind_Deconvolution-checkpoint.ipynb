## 1. Heatmap with varying hyperparameters

**Evaluate BL with number of inner loop iterations *i_psf, i_img = [1,....,5]* and outer iterations *i_rl = [1,...,10]***

```python
# TODO: How many images to include? simply mean?
# TODO: How to determine Sigma?

args = {}
args['data_path']= ''
args['source_folder']= '..\\Registered\\Heatmap'
args['target_folder']= ''
args['result_path'] = '..\\Blind_RL\\Heatmap'
args['psf'] = "C:/Users/Johan/Documents/FAU_Masterarbeit/Implementation/Data/PSF"

max_iter_inner = 5
max_iter_outer = 10
sigma = 1
```

*Obtain results*

```python
blind_rl = deconv.BlindRL(args)
```

```python
for i_psf in range(1,max_iter_inner+1):
    blind_rl.predict(args['source_folder'], max_iter_outer, max_iter_inner, i_psf, sigma, eval_img_steps = True)
```

*Create data structure with all results*

```python
# Load results from file
iter_rl= np.tile(np.repeat(np.arange(1,max_iter_outer+1), (max_iter_inner)), max_iter_inner)
iter_img= np.tile(np.arange(1,max_iter_inner+1), max_iter_inner*max_iter_outer)
iter_psf= np.repeat(np.arange(1,max_iter_inner+1), max_iter_inner*max_iter_outer)


df = pd.DataFrame()
df['Iter_RL'] = iter_rl
df['Iter_IMG'] = iter_img
df['Iter_PSF'] = iter_psf

df['snr'] = np.zeros((len(df),1))
step = max_iter_inner*max_iter_outer

for idx in range(0,len(df), step):
    row = df.iloc[[idx]]
    name = os.path.join(args['result_path'], 'results_' + str(max_iter_outer) +'_' +str(max_iter_inner)+'_'+ str(row['Iter_PSF'].values[0]) +'_' +str(sigma) +'.pkl')    

    with open(name, 'rb') as file:
        res = pickle.load(file)
    mice = list(res.keys())[5:] # iter_outer, _image, _psf, sigma, runtime, mouse 1 ...
    val = [res[k]['snr_img_steps'] for k in mice]
    val_np = np.array(val)
    val_np= val_np[np.newaxis,:] if len(val_np.shape) < 2 else val_np
    vals = np.mean(np.array(val), axis = 0)
    df['snr'].iloc[idx:idx+step] = vals
```

# Reconstruction error

Based on the findings above i_psf = , i_img = yields the best results. Below it is shown how the PSNR evolves over the amount of RL iterations

*Obtain results*

```python
blind_rl.predict(args['source_folder'], 10, 5, 3, 2, eval_img_steps = False, save_intermediate_res= True, planes_padding=10)
```

*Compute metrics*

```python
metrics = iqm.ImageQualityMetrics()

img_name = 'Alessandro_427_ArcCreERT2_Thy1GFP_Ai9_TRAP_2019-08-31_A2.tif'
psf_name = 'Alessandro_427_ArcCreERT2_Thy1GFP_Ai9_TRAP_2019-08-31_A2_psf.tif'
path = 'C:/Users/Johan/Documents/FAU_Masterarbeit/Implementation/Blind_RL'

original_path = 'C:/Users/Johan/Documents/FAU_Masterarbeit/Implementation/Registered/Raw'
org = tif.imread(os.path.join(original_path, img_name))
org_sm = blind_rl.preprocess(org, 2)

result = []

for k in range(0,11):
    i_n = 'iter_' + str(k) + img_name
    p_n =  'iter_' + str(k) + psf_name

    img = tif.imread(os.path.join(path,i_n))
    psf = tif.imread(os.path.join(path, p_n))
    convolved = convolve(img, psf, mode='same')
    result.append(metrics.compute_all(org_sm, convolved))
```

*Plot result*

```python
import matplotlib as mat
mat.rc_file_defaults()
```

```python
met_ssim = [x['ssim'] for x in result]

plt.plot(np.arange(0,11), met_ssim)
plt.xlim(0,10)
plt.ylim(0,1.1)
plt.grid()
plt.xlabel( 'Iterations')
plt.ylabel( 'SSIM')

plt.title('SSIM')
```

```python
met_psnr = [x['psnr'] for x in result]

plt.plot(np.arange(0,11), met_psnr)
plt.xlim(0,10)
plt.ylim(0,37)
plt.title('PSNR')
plt.xlabel( 'Iterations')
plt.ylabel( 'PSNR')

plt.grid()
```

```python
met_psnr
```

```python
met_mse = [x['mse'] for x in result]

plt.plot(np.arange(0,11), met_mse[0:])
plt.xlim(0,10)
plt.grid()
plt.yscale('log')
plt.xlabel( 'Iterations')
plt.ylabel( 'MSE')

plt.title('MSE')
```
